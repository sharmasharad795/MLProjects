Solution to 1(c)(iii) as below:

We import the math library here to implement the floor function.

We create two lists N_values,optimal_testerror_set in which we'll store values to represent the learning curve.
For the first step, we break the train data set into rows with class 0 and class 1.

```python
import math 

N_values=[]
optimal_testerror_set=[]
subset_tr_cl1=train_data[train_data['class']==1]
subset_tr_cl0=train_data[train_data['class']==0]
```

We run a for loop to break the training data set into sizes of {10,20,30,..200}. Then for each N, we calculate the no of rows belong to class 1 and class 0 which will form the new training set.

For each N , we then split the new training data into the dependent and independent variables' dataframes.


Afterwards, for each N, we run KNN algorithm with k values from {1,6,..N}. We find the test error for all suck K's and append it to list test_error_subset 

For each value of N, we create a dataframe which has the test errors and train errors corresponding to the K values.

After sorting our created dataframe by test error values, we append the least error values TO 'optimal_testerror_set' for the optimal K for each N 

We now plot these least test errors to the N sizes of training data to give us the learning curve.

Note: we have chosen the optimal error rate for each N as the one which is the least error rate for that particular N.
The ouput below shows the optimal K for each N size training data set


```python
for N in range(10,211,10):
    N_values.append(N)
    nrows_cl0=math.floor(N/3)
    nrows_cl1=N-nrows_cl0
    Nsubset_tr_cl1 = subset_tr_cl1[:nrows_cl1 ]
    Nsubset_tr_cl0=subset_tr_cl0[:nrows_cl0 ]
    Nsubset_tr=Nsubset_tr_cl0.append(Nsubset_tr_cl1,ignore_index=True)
 #   print(Nsubset_tr)
 #   print('\n')
 
    X_temp_trainset=Nsubset_tr.iloc[:,0:6]
    Y_temp_trainset=Nsubset_tr.iloc[:,6]
    k_values_subset=[]
    test_error_subset=[]
    train_error_subset=[]
    
    for k in range(1,N,5):
        k_values_subset.append(k)
        knn_classification=KNeighborsClassifier(n_neighbors=k,metric='minkowski', p=2)
        knn_classification.fit(X_temp_trainset,Y_temp_trainset)
        test_error_subset.append(1 - (knn_classification.score(X_testset,Y_testset)))
        train_error_subset.append(1 - (knn_classification.score(X_temp_trainset,Y_temp_trainset)))

    error_plot_subset=pd.DataFrame({'test_error_subset':test_error_subset,'train_error_subset':train_error_subset,'k_values_subset':k_values_subset})
    error_plot_subset_te=error_plot_subset.sort_values(by=['test_error_subset'])
    error_plot_subset_tr=error_plot_subset.sort_values(by=['train_error_subset'])
 #   print(error_plot_subset_tr)
    print(error_plot_subset_te)
    print("the optimal K for training set of size ",N, "is: ", error_plot_subset_te['k_values_subset'].iloc[0])
    print('\n')
    
    optimal_testerror_set.append(error_plot_subset_te['test_error_subset'].iloc[0])
```

```python
print("optimal k's are:",optimal_testerror_set)
```

```python
plt.plot(N_values,optimal_testerror_set)
plt.xlabel('Size of training set')
plt.ylabel('Best Test Error')
plt.show() 
```
